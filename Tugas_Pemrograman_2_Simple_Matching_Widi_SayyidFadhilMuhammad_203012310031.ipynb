```python
import networkx as nx
import matplotlib.pyplot as plt

# Fungsi untuk membaca input dan membuat graf
def create_graph(n, edges):
    G = nx.Graph()
    G.add_nodes_from(range(1, n+1))
    for edge in edges:
        G.add_edge(edge[0], edge[1])
    return G

# Fungsi untuk menemukan pencocokan maksimal dalam graf
def find_maximum_matching(G):
    matching = nx.algorithms.matching.max_weight_matching(G)
    return list(matching)

# Fungsi untuk mencetak output sesuai format yang diminta
def print_output(matching):
    print("=== Jumlah pasangan yang ditemukan ===")
    p = len(matching)
    print(p)
    print("")
    for pair in matching:
        print(pair[0], pair[1])

# Membaca input
print("=== Input jumlah vertex dan edges (n, m) ===")
n, m = map(int, input().split())
edges = []
print("")
print("=== Input pasangan verteks ===")
for _ in range(m):
    u, v = map(int, input().split())
    edges.append((u, v))

# Membuat graf
graph = create_graph(n, edges)

# Mencari pencocokan maksimal
maximum_matching = find_maximum_matching(graph)

# Mencetak output
print_output(maximum_matching)

# Membuat posisi node agar terlihat lebih baik
pos = nx.spring_layout(graph)

# Mengatur warna untuk node dan edge
node_colors = ['skyblue' if node not in maximum_matching else 'salmon' for node in graph.nodes()]
edge_colors = ['black' if (u, v) not in maximum_matching and (v, u) not in maximum_matching else 'red' for u, v in graph.edges()]

# Membuat visualisasi graf dengan warna yang berbeda untuk setiap pasangan yang terhubung
plt.figure(figsize=(6, 6))
nx.draw(graph, pos, with_labels=True, node_size=500, node_color=node_colors, edge_color=edge_colors, font_weight='bold', font_size=12, width=2)
plt.title('Graph Mathching')
plt.show()
```
